2 Describe the role of predefined keywords in Python and provide examples of how they are used in a program.

Ans

Keywords are reserved words in Python that have special meanings and cannot be used as identifiers (variable names, function names, etc.). 
Examples of keywords include if, else, while, for, def, class, import, try, except, return, True, False, None, etc.
Keywords are case-sensitive and cannot be redefined or overridden within a Python program.

i. if: Used to make decision
    if age >= 18:
        print("You are an adult.")

ii. for: Used to repeat actions a certain number of times or for each item in a list.
    for item in shopping_list:
        print(item)

iii. None: Represents the absence of a value or a null value.
    value = None
    
iv. while: Used to create a loop that keeps running as long as a condition is true.
    count = 0
    while count < 5:
        print(count)
        count += 1
v. in: Used to check if a value is present in a list, tuple, dictionary, or string.
    fruits = ["apple", "banana", "cherry"]
    if "banana" in fruits:
        print("Banana is in the list!")

Mutable Objects
Mutable objects are objects whose values can be changed after they are created. Think of them like a whiteboard you can write on, erase, and rewrite.

list = [1, 2, 3]

Immutable Objects
Immutable objects are objects whose values cannot be changed once they are created. Think of them like a printed book—you can't change the words on the pages without creating a new page.

Examples of Immutable Objects: String , tuple

5 Explain the concept of type casting in Python with examples.

Ans

Type casting, also referred to as type conversion, is the process of changing one data type to another in Python. 
Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.
Type casting is often necessary for performing arithmetic operations, data manipulation, and input/output operations in Python programs.

Types of Type Casting
Implicit Type Casting
Explicit Type Casting

i. Implicit Type Casting
Python sometimes converts data types automatically. This usually happens when performing operations with different types.

eg:
a = 5       # Integer
b = 2.5     # Float
c = a + b   # Python converts 'a' to a float and adds it to 'b'
print(c)    # Output: 7.5

ii. Explicit Type Casting
You explicitly convert data types using built-in functions like int(), float(), and str().

eg:
x = "10"
y = int(x)  # Converts string "10" to integer 10
print(y)    # Output: 10
print(type(y))  # Output: <class 'int'>


6 How do conditional statements work in Python? Illustrate with examples

Ans

Conditionals in Python are used to execute code based on the evaluation of one or more conditions.
Python supports conditional statements such as if, elif (else if), and else to control the flow of execution.
Conditionals can be nested to handle multiple conditions and control different branches of execution based on the outcome of logical expressions.

i. if Statement: Checks if a condition is true. If it is, the code block inside the if statement is executed.
eg:
age = 18
if age >= 18:
    print("You are an adult.")

ii. elif Statement: Short for "else if," it checks another condition if the previous if or elif conditions are false.
eg:
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D or lower")


iii. else Statement: Executes a block of code if none of the previous conditions are true.
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")
